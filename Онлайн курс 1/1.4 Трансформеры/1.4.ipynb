```python
from transformers import BartTokenizer, BartForConditionalGeneration

reviews = [
    "When we first watched it, we admired the visualization of the Gargantua black hole, "
    "but today the film is perceived as a very intimate, almost chamber story, where the "
    "boundless cosmos is just a backdrop for exploring the human heart.",
    
    "At the moment, this particular film can be called his loudest and boldest statement "
    "in cinema. Also, Interstellar fully demonstrates his philosophical research, the "
    "skills of a playwright and a visual master of special effects.",
    
    "Christopher Nolan's Interstellar is a grandiose space odyssey that combines "
    "scientific rigor and emotional depth.",
    
    "I believe that in the near future it is unlikely that anyone will be able to achieve such a level "
    "of quality and grandiosity of special effects – it is unforgettable, amazing, inspiring. It is "
    "impossible to describe the world of pleasure into which this painting immerses the viewer."
]

def summarize_reviews(reviews_list, model_name="facebook/bart-large-cnn"):

    combined_text = " ".join(reviews_list)

    tokenizer = BartTokenizer.from_pretrained(model_name)
    model = BartForConditionalGeneration.from_pretrained(model_name)

    inputs = tokenizer(
        combined_text, 
        max_length=1024, 
        truncation=True, 
        return_tensors="pt"
    )

    summary_ids = model.generate(
        inputs["input_ids"],
        max_length=130,
        min_length=30,
        num_beams=1,
        do_sample=False,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id
    )
    
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

result = summarize_reviews(reviews)
print("Краткое резюме:")
print(result)

#Краткое резюме:
#Christopher Nolan's Interstellar is a grandiose space odyssey that combines 
#scientific rigor and emotional depth. It is impossible to describe the world
#of pleasure into which this painting immerses the viewer.
```
