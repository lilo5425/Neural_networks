```python
N = 1024
ops = 2 * N**3
data = 12 * N**2

archs = [('GPU', 1.0, 15.0), ('TPU', 0.6, 3.0), ('NPU', 0.3, 0.0)]

results = []

for name, e_op, e_mem in archs:
    comp = ops * e_op
    mem = data * e_mem
    results.append((name, comp, mem))
    print(f"{name}: Вычисления={comp:.2e}, Память={mem:.2e}, Итого={comp+mem:.2e}")
```
