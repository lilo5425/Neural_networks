```python
from diffusers import StableDiffusionPipeline, EulerDiscreteScheduler
import torch

# Подключаем модель Stable Diffusion
model_id = "runwayml/stable-diffusion-v1-5"

# Загружаем пайплайн
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32)

prompt = "A majestic dragon flying over mountains at sunset, fantasy art"
pipe.scheduler = EulerDiscreteScheduler.from_pretrained(model_id, subfolder="scheduler")
# Генерируем изображение
image = pipe(
    prompt,
    num_inference_steps=15,
    guidance_scale=7.5,
    height=256,
    width=256,
).images[0]
image.save("dragon.png")
print("Изображение сохранено как 'dragon.png'")
```
